```python
nombre = input("Ingrese su nombre: ")
apellido = input("Ingrese su apellido: ")
#fecha
import datetime
#valores
pisos = 10
departamentos_por_piso = 4
departamentos_disponibles = [[True] * departamentos_por_piso for _ in range(pisos)]
compradores = []
precios = {
    'A': 3800,
    'B': 3000,
    'C': 2800,
    'D': 3500
}

#definiciones

def Elmenu():
    print("----------------")
    print("======MENÚ======")
    print("----------------")
    print("1- Comprar un departamento: ")
    print("2- Ver los departamentos disponibles: ")
    print("3- Ver el listado de compradores: ")
    print("4- Ver las ganancias totales: ")
    print("5- Salir: ")
    opcion = input("Seleccione una opción: ")
    return opcion

def comprar_depa():
    print("Departamentos disponibles:")
    ver_departamentos()
    piso = int(input("Ingrese el número de piso: "))
    tipo = input("Ingrese el tipo de departamento considerando las opciones como: A, B, C, D : ")
    departamento = tipo + str(piso)

    if not departamento_valido(departamento):
        print("Departamento inválido. Por favor, seleccione otro.")
        return

    if not departamento_disponible(departamento):
        print("El departamento seleccionado no está disponible. Por favor, seleccione otro.")
        return

    run = input("Ingrese el RUN del comprador (sin guion ni puntos): ")
    comprar_departamento_aux(departamento, run)
    print("La operación fue realizada correctamente.")

def comprar_departamento_aux(departamento, run):
    index = obtener_indice_departamento(departamento)
    departamentos_disponibles[index[0]][index[1]] = False
    compradores.append((departamento, run))

def ver_departamentos():
    for piso in range(1, pisos + 1):
        print(f"Piso {piso}: ", end="")
        for departamento in range(departamentos_por_piso):
            if departamentos_disponibles[piso - 1][departamento]:
                print(f"{chr(65 + departamento)}{piso}", end=" ")
            else:
                print("X", end=" ")
        print()

def ver_listado_compradores():
    compradores_ordenados = sorted(compradores, key=lambda x: int(x[1]))
    print("Listado de compradores:")
    for compra in compradores_ordenados:
        print(f"{compra[0]}: {compra[1]}")

def ver_ganancias_totales():
    total_por_tipo = {}
    for compra in compradores:
        tipo = compra[0][0]
        if tipo in total_por_tipo:
            total_por_tipo[tipo] += precios[tipo]
        else:
            total_por_tipo[tipo] = precios[tipo]

    total_general = sum(total_por_tipo.values())

    print("=====Ventas totales=====:")
    print("Tipo de Departamento\tCantidad\tTotal")
    for tipo, total in total_por_tipo.items():
        cantidad = total // precios[tipo]
        print(f"Tipo {tipo} {precios[tipo]} UF_____________{cantidad}____________{total} UF")

    print(f"TOTAL---> {len(compradores)} = {total_general} UF")

def departamento_valido(departamento):
    return departamento[0] in ['A', 'B', 'C', 'D'] and int(departamento[1:]) >= 1 and int(departamento[1:]) <= pisos

def departamento_disponible(departamento):
    index = obtener_indice_departamento(departamento)
    return departamentos_disponibles[index[0]][index[1]]

def obtener_indice_departamento(departamento):
    piso = int(departamento[1:]) - 1
    tipo = ord(departamento[0]) - 65
    return piso, tipo

def mostrar_salida():
    fecha_actual = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    print(f"Muchas gracias por usar el sistema, {nombre} {apellido}.\nFecha de salida: {fecha_actual}")

def main():
    while True:
        opcion = Elmenu()

        if opcion == "1":
            comprar_depa()
        elif opcion == "2":
            ver_departamentos()
        elif opcion == "3":
            ver_listado_compradores()
        elif opcion == "4":
            ver_ganancias_totales()
        elif opcion == "5":
            mostrar_salida()
            break
        else:
            print("La opción es inválida. Por favor, seleccione de nuevo una opcion valida.")

if __name__ == "__main__":
    main()


```
